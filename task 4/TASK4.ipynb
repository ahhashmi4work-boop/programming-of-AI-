```python
def is_safe(board, row, col, n):
    for i in range(row):
        if board[i] == col:
            return False
    i, j = row - 1, col - 1
    while i >= 0 and j >= 0:
        if board[i] == j:
            return False
        i -= 1
        j -= 1
    i, j = row - 1, col + 1
    while i >= 0 and j < n:
        if board[i] == j:
            return False
        i -= 1
        j += 1
    return True
def solve_n_queens(board, row, n, solutions):
    if row == n:
        solutions.append(board.copy())
        return
    for col in range(n):
        if is_safe(board, row, col, n):
            board[row] = col
            solve_n_queens(board, row + 1, n, solutions)
            board[row] = -1  
def n_queens(n):
    board = [-1] * n
    solutions = []
    solve_n_queens(board, 0, n, solutions)
    return solutions
n = int(input("Enter value of N: "))
solutions = n_queens(n)
print(f"\nTotal solutions for {n}-Queens: {len(solutions)}\n")
for sol in solutions:
    for i in range(n):
        row = ["Q" if sol[i] == j else "." for j in range(n)]
        print(" ".join(row))
    print()
```
