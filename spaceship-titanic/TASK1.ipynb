```python
from sklearn.preprocessing import LabelEncoder
import numpy as np

def convertObjToInt(cols):
    for col in cols:
        unique_vals = pd.DataFrame({col: df1[col].unique()})
        le = LabelEncoder()
        le.fit(np.ravel(unique_vals))
        df1[col] = le.transform(df1[col])

object_cols = df1.select_dtypes(include="object").columns.tolist()
convertObjToInt(object_cols)

```

```python
import pandas as pd
import numpy as np

def convert_bool_to_int64(df1):
    
    bool_cols = df1.select_dtypes(include=["bool"]).columns
    
    
    df1[bool_cols] = df1[bool_cols].astype("int64")
    
    return df1
df1 = convert_bool_to_int64(df1)
df1.info()
```

```python
import pandas as pd
import numpy as np

def convert_bool_to_int64(df1):
    
    bool_cols = df2.select_dtypes(include=["bool"]).columns
    
    
    df2[bool_cols] = df2[bool_cols].astype("int64")
    
    return df2
df2 = convert_bool_to_int64(df2)
df2.info()
```

```python
# Prepare training data - exclude PassengerId as it's just an ID
train_x = df1.drop(['Transported', 'PassengerId'], axis=1)
train_y = df1['Transported']

# Also prepare test data without PassengerId for prediction
df2_features = df2.drop('PassengerId', axis=1)
```

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.model_selection import cross_val_score
import numpy as np

# Try XGBoost first (usually better performance), fallback to RandomForest if not available
try:
    import xgboost as xgb
    # XGBoost model - typically achieves 80%+ accuracy
    model = xgb.XGBClassifier(
        n_estimators=300,
        max_depth=7,
        learning_rate=0.1,
        subsample=0.8,
        colsample_bytree=0.8,
        random_state=42,
        eval_metric='logloss',
        use_label_encoder=False
    )
    print("Using XGBoost model")
except ImportError:
    # Improved RandomForest as fallback
    model = RandomForestClassifier(
        n_estimators=300,
        max_depth=20,
        min_samples_split=5,
        min_samples_leaf=2,
        max_features='sqrt',
        random_state=42,
        n_jobs=-1
    )
    print("Using RandomForest model (install xgboost for better performance: pip install xgboost)")

# Train the model
model.fit(train_x, train_y)

# Cross-validation to check accuracy
cv_scores = cross_val_score(model, train_x, train_y, cv=5, scoring='accuracy')
print(f"\nCross-validation accuracy: {cv_scores.mean():.4f} (+/- {cv_scores.std() * 2:.4f})")
print(f"Individual CV scores: {cv_scores}")
print(f"Expected Kaggle score: ~{cv_scores.mean()*100:.2f}%")
```

```python
pred = model.predict(df2_features)
# Convert predictions to boolean (True/False) for Kaggle submission format
pred = pred.astype(bool)
```

```python
# Read original PassengerId from test.csv to preserve string format
df2_original = pd.read_csv('test.csv')
submission = pd.DataFrame({
    "PassengerId": df2_original["PassengerId"],
    "Transported": pred
})

submission.to_csv("submissions.csv", index=False)

```
